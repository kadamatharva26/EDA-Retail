```python
seg=df.Segment.value_counts().reset_index()
seg.columns=('SEGMENT','COUNT')
seg
```

```python
#pie chart
plt.pie(x='COUNT',labels='SEGMENT',data=seg, autopct='%1.0f%%')
```

```python
sns.countplot(x=df['Category'],hue=df['Region'])
```

```python
#shows sales data greater than 3500
h_sales=df[df['Sales']>3500]
h_sales
```

```python
#Graph for Sales data against State
plt.figure(figsize=(10,6))
sns.barplot(x=h_sales['Sales'],y=h_sales['State'],data=df)
```

```python
#gives maximum sales value
df.Sales.max()
```

```python
#gives minimum sales value
df.Sales.min()
```

```python
l_sales=df[df['Sales']<1]
plt.figure(figsize=(10,6))
sns.barplot(x=l_sales['Sales'],y=l_sales['State'],data=df)
```

```python
#shows data whose Profit value is less than 1
l_profit=df[df["Profit"]<1]
l_profit.head()
```
